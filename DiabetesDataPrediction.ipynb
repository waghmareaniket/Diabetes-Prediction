```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Diabetes.csv")
```

```python
data.head()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import svm
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
data['Outcome'].value_counts()
```

```python
x=data.iloc[:,:-1]
y=data['Outcome']
```

```python
sc=StandardScaler()
```

```python
sc.fit(x)
```

```python
sc_data=sc.fit_transform(x)
```

```python
x=sc_data
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=16)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#use svc because data is 0 and 1 in store
```

```python
svc=svm.SVC(kernel='linear')
```

```python
svc.fit(x_train,y_train)
```

```python
train_y_pred= svc.predict(x_train)
```

```python
test_y_pred=svc.predict(x_test)
```

```python
accuracy_score(y_train,train_y_pred)
```

```python
accuracy_score(y_test,test_y_pred)
```
